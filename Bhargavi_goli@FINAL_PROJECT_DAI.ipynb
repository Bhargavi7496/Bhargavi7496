```python
# DEPARTMENT-WISE HR EMPLOYEE ATTRITION
fig,ax = plt.subplots(figsize=(7,6))
sns.countplot(x = df["BusinessTravel"], hue= df["Attrition"],data=df)
plt.show()
```

```python
dep_attr = df.groupby(["Department"])["Attrition"].aggregate("count") # Grouping by departments 
dep_attr
```

```python
plt.figure(figsize=(5,4))
dep_attr.plot(kind='barh')
plt.xlabel('departemts')
plt.ylabel('count of attrited employees')
plt.title("department wise count of attrited employees")
plt.show()
```

```python
edu_attr = df.groupby(["EducationField"])["Attrition"].aggregate("count")
edu_attr
```

```python
plt.figure(figsize=(5,4))
edu_attr.plot(kind='barh')
plt.xlabel('Education_field')
plt.ylabel('count of attrited employees')
plt.title("department wise count of attrited employees")
plt.show()
```

```python
month = df.groupby(["Attrition"])["MonthlyIncome"].aggregate("sum")
month
```

```python
plt.figure(figsize=(5,4))
month.plot(kind='barh')
plt.xlabel('MonthlyIncome')
plt.ylabel('count of attrited employees')
plt.title("MonthlyIncome wise sum of attrited employees")
plt.show()
```
