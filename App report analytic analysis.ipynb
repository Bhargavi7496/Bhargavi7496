# Exploratory data analysis
 1. UNIVARIATE ANALYSIS

```python
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
plt.figure(figsize=(5,3))
plt.title('new users count')
sns.histplot(df['Content.New users'],bins=range(0,31,1))
```

```python
plt.figure(figsize=(5,4))
plt.title('Engaged sessions')
sns.boxplot(data=None,x=df['Content.Engaged sessions'],fliersize=3) # there are outliers at the both Q1 & Q2
```

```python
plt.figure(figsize=(5,4))         # No outliers
plt.title('Engagement rate')
sns.boxplot(data=None,x=df['Content.Engagement rate'],fliersize=5)
```

```python
plt.figure(figsize=(5,4))         
plt.title('Engaged sessions by users')
sns.boxplot(data=None,x=df['Content.Engaged sessions per user'],fliersize=5)
```

```python
plt.figure(figsize=(5,3))
plt.title('Average enagement time')
sns.histplot(df['Content.Average engagement time'],bins=range(0,31,2))
```

```python
plt.figure(figsize=(5,4))         # No outliers
plt.title('Content.Event count')
sns.boxplot(data=None,x=df['Content.Event count'],fliersize=5)
```

# 2. BIVARIATE ANALYSIS

```python
fig=plt.figure(figsize=(6,5))
sns.regplot(x = df['Content.Users'], y = df['Content.Average engagement time per session'], data=df,color="blue")
plt.show()
```

```python
fig=plt.figure(figsize=(5,5))
sns.regplot(x=df['Content.Average engagement time per session'],y=df['Content.Events per session'],data=data,color='blue')
plt.show()
```
